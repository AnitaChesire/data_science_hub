### **Boolean Masking**

This is a technique that overlays a boolean grid on the dataframe and selects the values in the dataframe that align with true values of the grid 

```python
data_frame = pd.DataFrame({'Planet': ['Mercury', 'Venus', 'Earth', 'Mars',
                   'Jupiter', 'Saturn', 'Uranus', 'Neptune'],
                   'radius_km' : [2440, 6052, 6371, 3390, 69911, 58232,
                     25362, 24622],
                     'moons': [0, 0, 1, 2, 80, 83, 27, 14],
                     'type': ['terrestrial', 'terrestrial', 'terrestrial', 'terrestrial',
                              'gas giant', 'gas giant', 'ice giant', 'ice giant'],
                              'rings': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes','yes'],
                              'mean_temp_c': [167, 464, 15, -65, -110, -140, -195, -200],
                              'magnetic_field': ['yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes']})
#reassign the name
planets = data_frame
planets
```

```python
planets.groupby(['type']).sum()
```

```python
planets.groupby(['type']).sum()[['moons']]
```

```python
planets.iloc[0:3,[3,5]]
```

```python
planets.loc[:,['Planet', 'moons']]
```

```python
#back to masking
mask = planets['moons'] <20
mask
```

```python
#apply the mask to the data frame
planets[mask]
```

```python
#another way to directly mask is 
planets[planets['moons'] < 20]
```

```python
#more conditional statements
# Create a Boolean mask of planets with fewer than 10 moons OR more than 50 moons.
fewer_mask = (planets['moons'] < 10) | (planets['moons'] > 50)

#apply the mask
planets[fewer_mask]
```

```python
# Create a Boolean mask of planets with more than 20 moons, excluding them if they
# have 80 moons or if their radius is less than 50,000 km.

more_mask = (planets['moons'] > 20) & ~(planets['moons'] == 80) & ~(planets['radius_km'] > 50000)
#apply the mask
planets[more_mask]
```

```python
planets.groupby(['type', 'magnetic_field']).agg(['mean', 'median'])
```

### **Merging Data**

There are two ways to merge data: using the concat or merge functio

```python
import numpy as np
import pandas as pd 

data_one = {
    "Planet":["Mercury", "Venus", "Earth", "Mars"],
    "radius_km": [2440, 6052, 6371, 3390],
    "moons": [0, 0, 1, 2],
}
data_two = {
    "Planet": ["Jupiter", "Saturn", "Uranus", "Neptune"],
    "radius_km": [69911, 58232, 25362, 24622],
    "moons": [80, 83, 27, 14]
}

df1 = pd.DataFrame(data_one)
df2 = pd.DataFrame(data_two)
df2
```

```python
#Merge data using concat
df3 = pd.concat([df1, df2], axis = 0)
df3
```

```python
#reset the indexes 
df3 = df3.reset_index(drop=True)
df3
```

```python
data = {'Planet': ['Earth', 'Mars','Jupiter', 'Saturn', 'Uranus',
                   'Neptune', 'Janssen', 'Tadmor'],
        'type': ['terrestrial', 'terrestrial','gas giant', 'gas giant',
                 'ice giant', 'ice giant', 'super earth','gas giant'],
        'rings': ['no', 'no', 'yes', 'yes', 'yes','yes', 'no', None],
        'mean_temp_c': [15, -65, -110, -140, -195, -200, None, None],
        'magnetic_field': ['yes', 'no', 'yes', 'yes', 'yes', 'yes', None, None],
        'life': [1, 0, 0, 0, 0, 0, 1, 1]
        }
df4 = pd.DataFrame(data)
df4
```

```python
#merge horizontally

inner = pd.merge(df3, df4, on= "Planet", how= "inner")
inner
```

```python
outer = pd.merge(df3, df4, on = "Planet", how= "outer")
outer
```

```python
right = pd.merge(df3, df4, on= "Planet", how= "right")
right
```

```python
left = pd.merge(df3, df4, on= "Planet", how= "left")
left
```
